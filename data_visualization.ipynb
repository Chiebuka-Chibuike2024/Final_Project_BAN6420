## Data Visualization

```python
# Import Libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load and Display Dataset

```python
# Load the CSV data into a Dataframe
df = pd.read_csv('data.csv')

# Display the dataset
df
```

## Visualizing Age-Income Distribution with Matplotlib

```python
# Visualizing Ages with highest Income using matplotlib

# Group by age and calculate the average income
age_income = df.groupby('Age')['Total Income'].mean().sort_values(ascending=False)

# Plot the data
plt.figure(figsize=(10, 6))
age_income.plot(kind='bar')
plt.title('Ages with the Highest Income')
plt.xlabel('Age')
plt.ylabel('Average Total Income')
plt.xticks(rotation=45)


# Save and Export Chart for Client Usage
plt.savefig('age_income_chart.png')
```

## Visualizing Age-Income Distribution with Seaborn

```python
# Visualizing Ages with highest Income using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x="Age", y="Total Income", data=df)
plt.title('Ages with the Highest Income')

# Save and Export Chart for Client Usage
plt.savefig('age_income_chart2.png')
```

- From the above visualizations, it is clear that individuals with higher incomes tend to be older. This trend is realistic, as income levels generally increase with age due to factors like career advancement, exposure and experience.





## Visualizing Gender Distribution Across Spending Categories

```python
# Gender distribution across spending categories

# Calculate the total spending by category for each gender
gender_expenses = df.groupby('Gender')[['Utilities', 'Entertainment', 'School Fees', 'Shopping', 'Healthcare']].sum()

# Plot the data
gender_expenses.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gender Distribution Across Spending Categories')
plt.xlabel('Gender')
plt.ylabel('Total Spending')
plt.legend(title='Expense Categories')

# Save and Export Chart for Client Usage
plt.savefig('gender_expenses_chart.png')
```

#From the above clustered chart, it is clear that school fees is the major expense category across both genders (female and male). Shopping is the least expense category across both genders as seen above.
